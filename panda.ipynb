```python
data = {'A':[2,5,np.nan,8,np.nan,9],
        'B':[np.nan,45,np.nan,89,63,np.nan],
        'C':[np.nan,74,np.nan,np.nan,85,4],
        'D':[10,20,30,40,50,60]}
```

For Handle missing values.....
1. Remove the Records.    2. Fill the Records.

```python
# removing....
df2.dropna()    # by default remove all those rows that containing missing values
```

```python
df2.dropna(axis=1)  # column
```

```python
print("Total missing values in your df: ",df2.isnull().sum().sum())
```

```python
df2.shape[0]
```

```python
df2.isnull().sum()/df2.shape[0]*100        # percentage amount
```

```python
# Filling......
# 1.numerical     2. categoricat
# numerical :-       # categoricat
# 1. constant          1.
# 2. mean avg
# 3. median
```

```python
df2.fillna(500)  # for permanent cahnges "inplace = True"
```

```python
df2['A'].fillna(500)
```

```python
df2['A'].mean()
```

```python
df2['A'].fillna(df2['A'].mean())
```

```python
df2['A'].fillna(df2['A'].median())
```
