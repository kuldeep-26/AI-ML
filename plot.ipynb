<h3>Line Plot:-</h3>   

    Line plots are a fundamental type of data visualization used in various fields for a variety of purposes like trend Analysis.
    Comparative Analysis, Highlighting Correlations, Forecasting,Visualizing Function Data, Data Exploring.


```python
# importing matplotlib library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
print("Libraries imported")
```

```python
country_population = [700,750,740,730,752,800,790]
country_population
```

```python
# Creating a Line Plot.....
plt.plot(country_population)
plt.show()
```

<h3>Markers in Matplotlib are used to highlight individual data points on a plot.</h3>

'o' = circle

'.' = point

's' = square

'x' = X

'd' = diamond

'^' = triangle

'<' = triangle left

'>' = triangle right

'+' = plus

'p' = pentagon

'h' = hexagon

```python
country_population = [700,750,740,730,752,800,820]
x_axis = [100,200,250,300,240,400,390]
plt.plot(x_axis,country_population,marker = 'o',ms = 18,mfc = 'yellow',color = 'red',ls = 'dotted',lw = 4)
plt.title("Lineplot Demo")
plt.xlabel('Axis-x')    # xlabel is used to label a name for x-axis....
plt.ylabel('country_population')     # ylabel is used to label a name for y-axis....
plt.show()
# ms = marker size...
# mfc = marker face colour....
# color = line color.....
# ls = line style....
# lw = line width....
```

# Bar plot:

Bar plots are a versatile and widely used type of plot in data visualization for displayong categorical data with rectangular bars.

Each bar's length or height is proportional to the value it represents. Bar plots can be vertical or horizontal and are useful for comparing different groups or tracking changes over time.

```python

country_population = [450,730,900,450,800]
country_name = ['japan','bangladesh','india','nepal','china']

# Creating a bar plot....
plt.bar(country_name,country_population)
plt.show()
```

```python
no_of_count = list(df['brand'].value_counts().values)
brand_name = list(df['brand'].value_counts().keys())
plt.bar(brand_name,no_of_count)
plt.xticks(rotation = 90)
plt.show()
```

# Histogram

A histogram is a graphical representation of the distribution of numerical data, providing insights into various aspects of the data.

Visualizing Data distribution, Identifying Patterns, Detecting Outliers, Summarizing Large Datasets etc.

```python
age = np.random.randint(1,100,500)
bin = [0,18,30,50,70,90,100]
# Bins paratmeter is used to define how the data in the age variable is grouped and displayed in the histogram

# Creating a histogram...
plt.hist(age,bins=bin,ec='k')
plt.show()
```

```python
# If the bins parameter is not passed in the plt.hist function, Matplotlib will use a method to determine the no. 
# and size of the bins.
plt.hist(age,ec='k')
plt.title("Histo-Graph for Profit & Time")
plt.xlabel('Travel Time (in min)')    # xlabel is used to label a name for x-axis....
plt.ylabel('DTC-Profit (in %)')
plt.show()
```

# Pie Plot:-

A pie plot, or pie chart is a circular statistical graphic divided into slices to illustrate numerical proportions.

Showing proportioms, Simplifying Data presentation, Highlighting Major Componenets, Visual Appeal etc.

```python
country_population = [450,730,900,450,800]
country_name = ['japan','bangladesh','india','nepal','china']

# Creating a pie plot...
plt.pie(country_population,labels = country_name,autopct = '%1.2f%%')
plt.show()
```

```python
country_population = [450,730,900,450,800]
country_name = ['japan','bangladesh','india','nepal','china']
explosion = [0,0,0.2,0,0]
# The autopct parameter in a pie chart is used to display the percentage values on the pie chart slices.
# The explode parameter in a pie chart is used to highlight or "explode" one or more slices from the rest of the pie.
plt.pie(country_population,labels = country_name,autopct = '%2.2f%%',explode = explosion)
plt.show()
```

```python
plt.figure(figsize=(6,8))     # width & height....
country_population = [450,730,900,450,800]
country_name = ['japan','bangladesh','india','nepal','china']
explosion = [0,0,0.2,0,0]
plt.pie(country_population,labels = country_name,autopct = '%2.2f%%',explode = explosion,wedgeprops = {'ec':'black','lw':3},shadow=True)
plt.show()
# The wedgeprops parameter in matplotlib's plt.pie function is used to customize the appearance of the wedges(slices) in a pie chart.
# It allows you to specify various properties of the wedges, such as their edges color, width, and line style.
```

# Box plot:-

A box plot, or box-and-whisker plot, is a graphical representation of the distribution of a dataset based on five key summary statistics:

The minimun

First Quartile(Q1)

Median(Q2)

Third Quartile(Q3)

The Maximum

It is used for identifying outliers

```python
df = pd.read_csv("Used_Bikes.csv")
df.drop_duplicates(inplace = True)
ktm = df[df['brand'] == "KTM"]

# Creating a Box plot
plt.boxplot(ktm['price'])
plt.show()
```
