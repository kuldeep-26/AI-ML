```python
cat_col = df.select_dtypes(include='O')
cat_col.head()
```

```python
num_col = df.select_dtypes(exclude='O')
num_col.head()
```

```python
cat_col['owner'].value_counts()
```

```python
dt = {'First Owner':1,'Second Owner':2,'Third Owner':3,'Fourth Owner Or More':4}
```

```python
df['owner'] = df['owner'].map(dt)    # single time execute
```

```python
df['owner']
```

```python
df['brand'].value_counts()
```

```python
bike_brand = list(df['brand'].value_counts().head(16).index)
bike_brand
```

```python
#(df['brand'] == "Bajaj") | (df['brand'] == "Royal Enfield")
df2 = df[df['brand'].isin(bike_brand)]
df2
```

```python
dt2 = {}
count = 1
for i in bike_brand:
    dt2[i] = count
    count += 1
dt2
```

```python
df2['brand'] = df2['brand'].map(dt2)
```

```python
df2
```

```python
df2.drop(['bike_name','city'],axis = 'columns',inplace = True)
```

Splitting data independent and dependent variables

```python
x = df2.drop(['price'],axis = 'columns')
y = df2[['price']]
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)  # test size = 20%
```

```python
# importing  linear regression algorithm class
from sklearn.linear_model import LinearRegression
```

```python
# instance creation
lnr = LinearRegression()
```

```python
# algorithm train
lnr.fit(x_train,y_train)
```

```python
# Model evaluation
print("Training Score: ",lnr.score(x_train,y_train) * 100)
print("Testing Score: ",lnr.score(x_test,y_test) * 100)
```

```python
x_test.shape
# total 1462 bikes
```

```python
# prediction or testing phase
prediction = lnr.predict(x_test)
```

```python
prediction.shape
```

```python
compare_df = pd.concat([x_test,y_test],axis=1)
```

```python
compare_df['prediction'] = prediction
```

```python
compare_df.head()
```

```python
compare_df['residual'] = compare_df['price'] - compare_df['prediction']
```
