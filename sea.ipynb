```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries imported")
```

# Barplot()

```python
df = pd.read_csv('Used_Bikes.csv')
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
brand_name = list(df['brand'].value_counts().keys())
brand_count = list(df['brand'].value_counts().values)
```

```python
plt.bar(brand_name,brand_count)
plt.xticks(rotation = 90)
plt.show()
```

```python
df = sns.load_dataset('tips')
df.head()
```

1. Relplot() - both column numerical [line & scatter]
2. Catplot() - 1 column numerical and 1 column categorical [strip,swarm,box,voilin,boxen,point,bar,count]

```python
sns.relplot(x = 'tip' , y = 'total_bill' , data = df,hue='day',style='sex')
plt.title('Scatter Plot for Time & Profit')
plt.xlabel('Profit')
plt.ylabel("Time-Travel")
plt.show()
```

```python
sns.relplot(x = 'tip' , y = 'total_bill' , data = df , hue = 'sex')
plt.show()
```

```python
sns.relplot(x = 'tip' , y = 'total_bill' , data = df , hue = 'sex' , style = 'smoker')
plt.show()
```

```python
sns.relplot(x = 'tip',y = 'total_bill',data = df,hue = 'sex',style = 'smoker',col = 'day') 
plt.show()
```

```python
sns.relplot(x = 'tip',y = 'total_bill',data = df,hue = 'time',col = 'day') 
plt.show()
```
